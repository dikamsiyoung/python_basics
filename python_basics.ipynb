```
rainfall_mi = "1.65, 1.46, 2.05, 3.03, 3.35, 3.46, 2.83, 3.23, 3.5, 2.52, 2.8, 1.85"
rainfall_list = rainfall_mi.split(',')

num_rainy_months = 0
for measurement in rainfall_list:
    measurement_num = float(measurement)
    if measurement_num > 3.0:
        num_rainy_months += 1
        
print('Number of months with greater than 3.0: ', num_rainy_months)
```

```
sentence = "students flock to the arb for a variety of outdoor activities such as jogging and picnicking"

sentence_list = sentence.split(' ')
same_letter_count = 0
for word in sentence_list:
    if word[0] == word[-1]:
        same_letter_count += 1

print('Words in sentence with the same letters: ', same_letter_count)
```

```
items = ["whirring", "wow!", "calendar", "wry", "glass", "", "llama","tumultuous","owing"]

acc_num = 0
for word in items:
    if 'w' in word:
        acc_num += 1

print("Words in item with 'W': ", acc_num)
```

```
sentence = "python is a high level general purpose programming language that can be applied to many different classes of problems."

sentence_list = sentence.split(' ')
num_a_or_e = 0
for word in sentence_list:
    if 'a' in word or 'e' in word:
        num_a_or_e += 1

print("Words in sentence with 'a' or 'e': ", num_a_or_e)
```

```
s = "singing in the rain and playing in the rain are two entirely different situations but both can be fun"
vowels = ['a','e','i','o','u']

# Write your code here.
num_vowels = 0
for char in s:
    for vowel in vowels:
        if vowel == char:
            num_vowels += 1

print("Number of vowels in s: ", num_vowels)
```
